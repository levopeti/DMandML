### Linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
x_train_simple = np.load("./data/x_train_simple.npy")
y_train_simple = np.load("./data/y_train_simple.npy")
```

```python
lin_reg = LinearRegression()
```

```python
p = lin_reg.fit(x_train_simple, y_train_simple)
```

```python
alpha = p.coef_
```

```python
beta = p.intercept_
```

```python
alpha, beta
```

```python
plt.figure()
plt.plot(x_train_simple, y_train_simple, "ro")
plt.plot(x_train_simple, alpha * x_train_simple + beta, "k", linewidth=3)
plt.grid()
plt.show()
```

## Logistic regression

```python
x_train_logistic = np.load("./data/x_train_logistic.npy")
y_train_logistic = np.load("./data/y_train_logistic.npy")
```

```python
log_reg = LogisticRegression(solver="lbfgs")
```

```python
p = log_reg.fit(x_train_logistic, y_train_logistic)
```

```python
def model(x):
    return (1 / (1 + np.exp(-x)))
```

```python
x = np.linspace(0.1, 2.05, 300)
est = model(x * p.coef_ + p.intercept_).ravel()
```

```python
plt.figure()
plt.plot(x_train_logistic, y_train_logistic, "ro")
plt.plot(x, est, "k", linewidth=3)
plt.xlabel("Hours studying", fontsize=15)
plt.ylabel("Probability of passing the midterm", fontsize=15)
plt.grid()
plt.show()
```
