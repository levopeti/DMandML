```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.tree import plot_tree as pt
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data/wineset.csv') # Reading the data set
df = pd.get_dummies(df) # Convert the categorical features into numerical ones
df = df[['Alcohol_content', 'Sweetness_dry', 'Sweetness_semi-sweet', 'Sweetness_sweet', 'Type_red', 'Type_rose', 'Type_white', 'Popular']] # Rearranging the data

## Split the data into features and targets

Y_train = df.iloc[:, [-1]]
X_train = df.drop(Y_train.columns, axis=1)
```

```python
clf = DTC(criterion='entropy', random_state=42) # Creating the classifier object

clf.fit(X_train, Y_train) # Using the fit() class method to fit the model to the available data

Y_pred = clf.predict(X_train) # Prediction with the trained model
Y_pred_proba = clf.predict_proba(X_train) # Prediction probabilities with the trained model
```

```python
fig = plt.figure(figsize=(15, 10)) # Initiating a figure with a predetermined size
_ = pt(clf, feature_names=X_train.columns, class_names=['No', 'Yes'], filled=True) # Use the native tree plotter function of sklearn
```
