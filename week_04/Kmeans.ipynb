```python
# Importing necessary libraries

import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import MiniBatchKMeans, KMeans
from sklearn.datasets import make_blobs
```

```python
# Create samples for clustering

n_samples = 1500 # --> sample size
random_state = 42 # --> seed for the random number generation used in cluster centroid initialization
X, y = make_blobs(n_samples=n_samples, random_state=random_state, centers=3) # --> generate data to be clustered
```

```python
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X) # --> create clustering model as object
#y_pred = KMeans(n_clusters=3, random_state=random_state, init='random').fit_predict(X)

# Plot results

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1]) # -->  raw data scatter plot
plt.title("Incorrect Number of Clusters")
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y_pred) # --> coloring based of the scatterplot on prediction
plt.show()
```

```python
# Data generation for the data size problem

n_samples = 2000000
random_state = 42
X, y = make_blobs(n_samples=n_samples, random_state=random_state, centers=3)
```

```python
# Showing the difficulties of Kmeans with large sample size

y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X)

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1])
plt.title("Incorrect Number of Clusters")
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
# MiniBatchKMeans for easing the pressure of large sample size

y_pred = MiniBatchKMeans(n_clusters=3, random_state=random_state).fit_predict(X)

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1])
plt.title("Incorrect Number of Clusters")
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
# Showing the sensitivity of Kmeans to the number of clusters

n_samples = 1500
random_state = 42
X, y = make_blobs(n_samples=n_samples, random_state=random_state, centers=3)

y_pred = KMeans(n_clusters=4, random_state=random_state).fit_predict(X)

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1])
plt.title("Incorrect Number of Clusters")
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
# Showing the sensitivity of Kmeans for anisotropic data

n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state, centers=3)
transformation = [[0.5, 0.5], [-0.4, -0.8]]
X_aniso = np.dot(X, transformation)

y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1])
plt.title("Anisotropicly Distributed Blobs")
plt.subplot(122)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
```

```python
# Difficulties of Kmeans for variance in variances

n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, cluster_std=[1.0, 3.5, 0.5], random_state=random_state, centers=3)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X)


plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1])
plt.title("Unequal Variance")
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```
